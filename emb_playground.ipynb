```python
import pandas as pd
import nltk
import numpy as np

encoding = "ISO-8859-1"
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
train = pd.read_csv('train.csv', usecols=['id', 'product_title'], encoding=encoding, index_col=['id'])
train.head()
```

```python
embeddings = {}
with open('emb/glove.42B.300d.txt') as f:
    for line in f:
        values = line.strip().split()
        embeddings[values[0]] = np.array([values[1:]], dtype=np.float32)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
```

```python
not_in_emb = []
for line in train.product_title.values:
    line = line.lower()
    tokens = tokenizer.tokenize(line.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    not_in_emb += [token for token in tokens if token not in embeddings]

for word in set(not_in_emb):
    print(word)
    
```
